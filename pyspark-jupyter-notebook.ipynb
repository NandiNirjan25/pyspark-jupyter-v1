# Complete PySpark Starter Code
Run the cell below to initialize Spark, create data, run transformations, and execute SQL.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc, avg

# 1. Initialize Spark Session
print("Initializing Spark Session...")
spark = SparkSession.builder \
    .appName("My_Single_Cell_PySpark_Project") \
    .master("local[*]") \
    .getOrCreate()

print(f"Spark Version: {spark.version}\n")

# 2. Create sample data
data = [
    ("Nirjan", "Engineering", 75000, 23),
    ("Alice", "Sales", 60000, 28),
    ("Bob", "Engineering", 80000, 35),
    ("Charlie", "HR", 55000, 42),
    ("Diana", "Sales", 65000, 29)
]
columns = ["Name", "Department", "Salary", "Age"]

print("--- Original DataFrame ---")
df = spark.createDataFrame(data, columns)
df.show()

# 3. Data Transformations
print("--- High Earners in Engineering (Salary > 60k) ---")
high_earners_eng_df = df.filter((col("Department") == "Engineering") & (col("Salary") > 60000))
high_earners_eng_df.show()

# 4. Aggregations
print("--- Average Salary by Department ---")
avg_salary_df = df.groupBy("Department") \
                  .agg(avg("Salary").alias("Average_Salary")) \
                  .orderBy(desc("Average_Salary"))
avg_salary_df.show()

# 5. Using Spark SQL
df.createOrReplaceTempView("employees")
sql_query = """
    SELECT Department, COUNT(*) as Employee_Count
    FROM employees
    WHERE Age < 30
    GROUP BY Department
"""
print("--- SQL Query Results (Employees under 30 per department) ---")
sql_results = spark.sql(sql_query)
sql_results.show()

# 6. Clean Up
print("Stopping Spark Session...")
spark.stop()
print("Spark Session stopped successfully.")
```
